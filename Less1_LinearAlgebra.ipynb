задание 6

Для умножения матриц представленных в задании, нужно выполнить умножение элементов строк первой матрицы на элементы столбцов второй матрицы.

Given:
A = [3, 4, -1]
B = [[5], [0], [7]]

To find C = A * B:
C = A[0]*B[0] + A[1]B[1] + A[2]B[2]
C = 35 + 40 + (-1)*7
C = 15 + 0 - 7
C = 8

Таким образом, сумма элементов главной диагонали получившейся матрицы C равна 8.
